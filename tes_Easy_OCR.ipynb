```python
import cv2
import easyocr
import os
from os import listdir

# get the path/directory
folder_dir = "testSample"
reader = easyocr.Reader(['ch_sim','en']) # this needs to run only once to load the model into memory
for images in os.listdir(folder_dir):
  
  result = reader.readtext('testSample/'+images)
  print(str(result) + str(images))

```

```python
import cv2
import pytesseract
import os
from os import listdir

# get the path/directory
folder_dir = "testSample"

for images in os.listdir(folder_dir):

  mobimg = cv2.imread('testSample/'+images)

#conversion to grayscale 
  mobimg = cv2.cvtColor(mobimg, cv2.COLOR_BGR2GRAY)

#thresholding 
  mobimg = cv2.threshold(mobimg, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]

#remove noise
  mobimg = cv2.medianBlur(mobimg, 5)


#read mobile number using OCR pytesseract
#pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'
  text = pytesseract.image_to_string(mobimg, lang='eng')

  print(text)  

```
